# Analysis and Visualisation: NBA Trends

In this project, weâ€™ll analyse and visualise data from  NBA (National Basketball Association) and explore possible associations.

Note: this dataset has been already preprocessed. To  make things more transparent and brief, there are only 5 teams included: **"Knicks"**, **"Celtics"**, **"Thunder"**, **"Spurs"**, **"Nets"**.

Columns description: 

- `game_id`: an identification number of a game;
- `year_id`: a year the game played;
- `fran_id`: a host's franchise identification name; 
- `opp_fran`: an opponent's franchise identification name;
- `game_location`: a code to mark game locations (`H` - home, `A` - away);
- `is_playoffs`: a code to mark whether the game is a playoff - a final game or a series of games played to break a tie (`1` - playoff, `0` - not a playoff);
- `pts`: host's points;
- `opp_pts`: opponent's points;
- `game_result`: a code to mark result of a game for a host (`W` - win, `L` - loss);
- `forecast`: chances for winning (closer to 0 - losing, closer to 1 - winning);
- `point_diff`: a difference in points between a host and an opponent.

## Analysing Relationships between Quantitative and Categorical

First, let's import necessary libraries and upload `nba_games.csv` to `nba` variable. 


```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, chi2_contingency
import matplotlib.pyplot as plt
import seaborn as sns


nba = pd.read_csv('nba_games.csv')
```

Let's start with smaller slice of the dataset: games of the year 2010.

```python
# Subset Data of 2010 
nba_2010 = nba[nba.year_id == 2010]
nba_2010.head(10)
```

```python
nba.groupby('game_id').size()[nba.groupby('game_id').size() > 1]
nba[nba.game_id == '194612070BOS']
```

Suppose we want to compare average number of points per game for **"Knicks"** and **"Nets"** in 2010.

```python
knicks_pts = nba_2010.pts[nba_2010.fran_id == 'Knicks']
knicks_pts
```

```python
nets_pts = nba_2010.pts[nba_2010.fran_id == 'Nets']
nets_pts
```

Let's check difference between their means.

```python
diff_means_2010 = knicks_pts.mean() - nets_pts.mean()
diff_means_2010
```

Result: **"Knicks"** have almost 10 points more on average than **"Nets"**.

Now we are going to visualise points of **"Knicks"** and **"Nets"** in 2010 as overlapping histograms to see distribution of their points compared to each other. 

```python
# Filter slice of a year 2010 so it contains only 2 teams
knicks_nets_2010 = nba_2010[(nba_2010.fran_id == 'Knicks') | (nba_2010.fran_id == 'Nets')]
```

```python
# Draw overlapping histplot 
def visualise_histplotoverlap(dataset, palette, title, xlabel, ylabel, legend_labels, legend_title):
    # Prepare grid
    plt.figure(figsize = (15,10))

    # Plot a histogram
    ax = sns.histplot(
        dataset, 
        x=dataset.pts, 
        hue=dataset.fran_id, 
        stat="count", 
        legend=True, 
        palette= palette)

    # Add descriptions
    ax.legend(title=legend_title, 
    title_fontsize='large', 
    loc='best', 
    labels=legend_labels, 
    fontsize = 'large')

    ax.set_title(title, fontsize = 'x-large')
    ax.set_xlabel(xlabel, fontsize = 'large')
    ax.set_ylabel(ylabel, fontsize = 'large')

    return plt.show()
```

```python
visualise_histplotoverlap(
    knicks_nets_2010, 
    palette= 'magma', 
    title='Points Distribution', 
    xlabel='Points', 
    ylabel='Games Count', 
    legend_labels=['Nets', 'Knicks'], 
    legend_title='Team Names')
```

Result: **"Knicks"** results are better in general, although **"Nets"** have more than 20 games with about 90 points. 

As a second step we are going to visualise points of **"Knicks"** and **"Nets"** in 2010 as boxplots to see distribution of their points compared to each other with median (black line on the boxes) and mean (black cross on the boxes). 

```python
# Draw boxplot
def visualise_boxplot(dataset, palette, title, xlabel, ylabel):
    plt.figure(figsize = (15,10))
    ax = sns.boxplot(
        x= dataset.fran_id, 
        y= dataset.pts,
        palette=palette, 
        showmeans=True,
        meanprops={"marker": "+",
            "markeredgecolor": "black",
            "markersize": "10"})

    # Add descriptions
    ax.set_title(title, fontsize = 'x-large')
    ax.set_xlabel(xlabel, fontsize = 'large')
    ax.set_ylabel(ylabel, fontsize = 'large')

    return plt.show()
```

```python
visualise_boxplot(
    knicks_nets_2010, 
    palette='magma', 
    title='Points Distribution', 
    xlabel='Team Names', 
    ylabel='Points')
```

Result: 
- On a boxplot it is more apparent that **"Knicks"** results in 2010 are better, because significant part of their points (the box itself) is located in the area of higher points `95-110`.
- The spread of general points for **"Knicks"** is observed from `~80` to `~130`, for **"Nets"** - from `~70` to `~120`
- Median and mean are close to each other on both plots.
- Each team has only one outlier, **"Knicks"** around `70` points, **"Nets"** - `~125`.


Now we will visualise points of **"Knicks"** and **"Nets"** via line plot over years.

```python
# Filter slice of over years so it contains only 2 teams
knicks_nets_all_years = nba[(nba.fran_id == 'Knicks') | (nba.fran_id == 'Nets')]
```

```python
# Draw lineplot function
def visualise_lineplot(dataset, palette, title, xlabel, ylabel, legend_title):
    # Prepare grid
    plt.figure(figsize = (15,10))
    # Create lineplot
    ax = sns.lineplot( 
        x=dataset.year_id, 
        y=dataset.pts, 
        hue=dataset.fran_id,  
        palette= palette,
        legend=True)

    # Add descriptions
    ax.legend(title=legend_title, 
    title_fontsize='large', 
    loc='best',
    fontsize = 'large') 

    ax.set_title(title, fontsize = 'x-large')
    ax.set_xlabel(xlabel, fontsize = 'large')
    ax.set_ylabel(ylabel, fontsize = 'large')

    return plt.show()
```

```python
visualise_lineplot(
    knicks_nets_all_years, 
    palette='Set2', 
    title='Points Fluctuations Over Years', 
    xlabel='Years', 
    ylabel='Points',  
    legend_title='Team Names')
```

Nevertheless they also have similar to **"Knicks"** decline with slightly better fluctuations `90-100` during the middle of the 90s to the middle of the 2000s.
- After 2005 **"Nets"** points continued their fluctuations but on the lower level around `90-100` points.

Let's see the same results but now each team on a separate plot. 

```python
# Draw faceplot
def visualise_facetplot(dataset, col, hue, palette, xcolumn, ycolumn, xlabelx, ylabel, legend_title):
    # Create grid
    g = sns.FacetGrid(
        data=dataset, 
        col=col, 
        col_wrap=3, 
        height=4, 
        aspect=1.5, 
        hue=hue, 
        palette=palette,
        sharex=False, 
        sharey=True,
        legend_out=True)
    
    # Add plots to grid
    g.map(sns.lineplot, xcolumn, ycolumn).add_legend(
        title=legend_title, 
        title_fontsize='large', 
        fontsize = 'medium')

    # Add labels
    g.set_axis_labels(xlabelx, ylabel)

    return plt.show()
```

```python
visualise_facetplot(
    nba,  
    col='fran_id',
    hue='fran_id', 
    palette='Set2', 
    xcolumn='year_id', 
    ycolumn='pts', 
    xlabelx="Years", 
    ylabel="Points",
    legend_title = 'Average Points n.\nover Years\n(with Min and Max)')
```

Note: results of **"Knicks"** and **"Celtics"** are skewed because, on one hand, they started playing much earlier than others and played more games, but on the other - they started from meagre points, compared to modern standards. That is because current rules, when a team has only 24 seconds for an attack, were set only in 1954 (https://dunkorthree.com/shot-clock-basketball/). Before that teams could "dance" around the field without scoring points, which was rather boring for viewers. 

Let's try to fix skewness to better understand whose results are relatively better among those five teams, not only in points but also over years. To mitigate this skewness we are going to compare the mean number of points each of the 5 teams got in all their games over 1 year (with min & max borders), not relative to each other, but rather relative to the average points number of all games of all teams per year. 

To calculate this mean value we are going to use not only the points of the 5 teams, that we already know, but also their opponents, mentioned in `opp_pts`. 

As a result, we'll see which team plays better or worse than average, even if the average itself was significantly lower in the past, and in general, fluctuates each year.

```python
# Find mean values across 'pts' and 'opp_pts'
mean_pts_per_year = nba.groupby('year_id').mean(['pts', 'opp_pts'])
mean_pts_per_year['mean_pts'] = mean_pts_per_year[['pts', 'opp_pts']].mean(axis=1)
mean_pts_per_year
```

```python
# Find percentages from mean for each game of 5 teams
nba['perc_from_mean'] = nba.apply(lambda x: x.pts / mean_pts_per_year.mean_pts[x.year_id] * 100 - 100, axis=1)
```

```python
# Draw FacetGrid
g = sns.FacetGrid(
    data=nba, 
    col='fran_id', 
    col_wrap=2, 
    height=4, 
    aspect=2, 
    hue='fran_id', 
    palette='Set2',
    sharex=False, 
    sharey=True,
    legend_out=True)

# Add main plots on FacetGrid
g.map(sns.lineplot, 'year_id', 'perc_from_mean').add_legend( 
    title='Average Points n.\nover Years\n(with Min and Max)', 
    title_fontsize='large', 
    fontsize = 'medium')

# Add additional plots on FacetGrid
def a(color, **kwargs):
    # Add mean line
    plt.axhline(0, color='purple')

    # Add border line signifying introduction of shot clock
    plt.axvline(1954, alpha=0.5)

    # Separate y axis
    plt.twinx()

    # Add secondary lineplot
    p = sns.lineplot(
        y=mean_pts_per_year.mean_pts,
        x=mean_pts_per_year.index,
        marker='o',
        **kwargs,
        legend=False
    )
    # Add label for y axis of secondary lineplot
    p.set_ylabel('Points')    

# Draw all plots
g.map(a)

# Add y and x labels of main plot
g.set_axis_labels('Years', 'Percentages from Mean(%)')
```

Result:
- All results correspond to what we've seen on the boxplots.
- Now we can see why **"Celtics"** were the best - they have been playing better than average level almost all their history. They started from very low points, `5% lower than standard 60-70 points` in the 40s, but in upcoming years improved their results and had only 2 periods when they played `3-5% lower 100-110 points` that were the averages in the late 70s and late 2010s.

- **"Spurs"** have a pretty impressive carrier too. They had 4 downs but these periods were relatively short and followed by significant rises. For instance, after a crisis in the 70s, they scored way above average (`5-10% above 100-110 points`) for many years till the second less significant down (`3-5% lower 110 points`) in the late 80s. The most significant crisis they had in the late 90s - `almost 10% below 100-90 average points` that coincided with the major crisis for all teams in the late 90s beginning of the 2000s.

- **"Thunder"** played mostly `3-10% below 110 points` (which was a really high average) in the 70s and 90s. After averages started lowering to 100-90 points, in the late 90s and further, their scores became the same 3-10% higher than averages. So one can tell their points were always relatively stable and can be even perceived as good considering the lowering of average scores in recent years. 

- **"Knicks"** and **"Nets"** have similar history of many periods of scoring `5-10% lower than averages` in different years. Their results can be perceived as worse than average most of the time, but with some shors bright moments of scoring `5% higher than 100-110 points`.

## Analyzing Relationships between Categorical variables

The variable `game_result` indicates whether a team won a particular game. The variable `game_location` indicates whether a team was playing at home or away. Do teams tend to win more games at home compared to away? Let's create a contingency table of frequencies and table of propotions to help us determine if these categorical variables are associated.

```python
# Create contingency table of frequencies
result_location_freq = pd.crosstab(nba_2010.game_result, nba_2010.game_location)
result_location_freq
```

```python
# Create contingency table of propotions
result_location_freq = pd.crosstab(nba_2010.game_result, nba_2010.game_location)
result_location_prop = result_location_freq/len(nba_2010)
result_location_prop
```

```python
# Find marginals for results and locations
result_marginals = result_location_prop.sum(axis=0)
location_marginals = result_location_prop.sum(axis=1)
result_marginals, location_marginals
```

In order to understand whether these questions are associated, we can use the marginal proportions to create a contingency table of expected proportions if there were no association between these variables, and compare with the actual one above. 

```python
# Find expected contingency table of frequencies
chi2, pval, dof, expected = chi2_contingency(result_location_freq)
result_location_prop_exp = np.round(expected)
result_location_prop_exp
```

```python
# Compare expected and actual tables
result_location_prop_exp, result_location_freq
```

Tables look somewhat different, so we can conclude that result and location are probably. Let's find how high with Chi-Square statistic to summarize how different these two tables are.

```python
chi2, pval, dof, expected = chi2_contingency(result_location_freq)
chi2
```

Result:
The interpretation of the Chi-Square statistic is dependent on the size of the contingency table. For a 2x2 table (like the one weâ€™ve been investigating), a Chi-Square statistic larger than around `4` would strongly suggest an association between the variables. We have Chi-Square around `6.5`, so variables are associated. 


## Analyzing Relationships Between Quantitative Variables

In the dataset there is a column that tells us the probability that each team will win the game. In the data, this is saved as `forecast`. The `point_diff` column gives the margin of victory/defeat for each team (positive values mean that the team won; negative values mean that they lost). Did teams with a higher probability of winning also tend to win games by more points?

Using nba_2010, we'll calculate the covariance between `forecast` and `point_diff`. 

```python
np.cov(nba_2010.forecast, nba_2010.point_diff)
```

Covariance is 1.37 which indicates rather small linear assiciation. Let's find correlation for these variables. 

```python
corr_forecast_point_diff, p = pearsonr(nba_2010.forecast, nba_2010.point_diff)
corr_forecast_point_diff
```

A correlation larger than about `.3` indicates a small positive linear association. We have `0.44`. Let's check it on a scatterplot.

```python
# Draw scatterplot
def visualise_scatterpot(dataset, title, xlabel, ylabel, color):
    # Prepare figure
    plt.figure(figsize = (15,10))
    # Draw plot
    plt.scatter(dataset.forecast, dataset.point_diff, color = color)
    # Add 0 line
    plt.axhline(0, color='grey')
    # Add descriptions
    plt.title(title, fontsize = 'x-large')
    plt.xlabel(xlabel, fontsize = 'large')
    plt.ylabel(ylabel, fontsize = 'large')
    
    return plt.show()

visualise_scatterpot(
    nba_2010, 
    title = 'Association between Forecast and Difference in Number of Points', 
    xlabel= 'Forecasted Win Probability', 
    ylabel='Point Differential (pos.=win, neg.=loss)', 
    color='salmon')
```

Indeed we have a positive linear association.
