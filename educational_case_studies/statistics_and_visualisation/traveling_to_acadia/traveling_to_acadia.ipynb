# Traveling to Acadia

The goal of this project is to inform customers of a travel agency about the best time to visit one of best destinations to see flowers in bloom - Acadia, Maine. We will use flower bloom data, and flight information to create visualisation and recommend the best time of year for someone to make a trip to Maine.

## Data Investigation

Let's start with necessary imports, as well as loading data from CSVs:

- `in_bloom.csv` is a NumPy array containing 900 numbers. Each number is a day of the year, which represents the average start date of a flower blooming.
- `flights.csv` is a NumPy array with 11,000 numbers. Each number is a day of the year, which represents a flight from your hometown to airports near Acadia, Maine.

Let's create two histograms using matplotlib, each displaying the frequency of an occurrence each day of the year (either flights or flower blooms). We will use the `in_bloom` variable to find a count of the number of flowers that start blooming each day of the year. We will use the `flights` variable to find a count of the number of flights that occur each day of the year.

```python
import numpy as np
from matplotlib import pyplot as plt

# Load in data
in_bloom = np.loadtxt(open("in-bloom.csv"), delimiter=",")
flights = np.loadtxt(open("flights.csv"), delimiter=",")
```

## Data Visualisation 

Let's use Matplotlib to create a flights histogram. We will set the range of our histogram to (0, 365) and set the number of bins in our plot to 365, so we have a separate bin for each day of the year.

```python
# Draw histplot
plt.hist(flights, range=(0, np.amax(flights)), bins = 365)
plt.xlabel('Number of Fligts')
plt.ylabel('Counts')
plt.title('Number of flights that occur each day of the year')
```

3.Time to add the second histogram that displays the number of flowers that begin to bloom each day of the year. We are going to display both histograms on a grid for better comparison.

```python
# Prepare figure
plt.figure(figsize=(16, 10))
plt.subplot(211)
# Draw 1st histplot
plt.hist(flights, range=(0, np.amax(flights)), bins = 365)
plt.xlabel('Number of Fligts')
plt.ylabel('Counts')
plt.title('Number of flights that occur each day of the year')

# Draw 1st histplot
plt.subplot(212)
plt.hist(in_bloom, range=(0, np.amax(in_bloom)), bins = 365)
plt.xlabel('Number of Flowers')
plt.ylabel('Counts')
plt.title('Number of flowers that in bloom each day of the year')
# Prevent the labels from overlapping with the graphs
plt.tight_layout() 
```

## Conclusion


Now let's think how would we use these histograms to help inform customers when they should go to Acadia, Maine. For example, if someone said they wanted to visit Acadia while there weren’t many people there, but while flowers were blooming, when would we recommend to go there?

Based on these histograms, we coul would propose they go between days 250 and 300, which is the Fall. The flight’s histogram shows fewer flights between these days, while the flowers histogram indicates there should be a number of flowers in bloom during this time.
