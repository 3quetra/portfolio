# Visualizing World Cup Data

The goal of this project is using Seaborn explore data from the Fifa World Cup from 1930-2014 to analyze trends and discover insights about football game.

This Fifa World Cup data is from Kaggle. Using Seaborn we will create a series of plots that explore aggregates and distribution across the goals scored in World Cup games.

## Data Investigation

First, let's explore the matches data to understand what information is present.

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

# Load WorldCupMatches.csv into a DataFrame 
df = pd.read_csv('fifa_matches.csv')
df.head()
```

The data in **fifa_matches.csv** has the goals scored in each match broken up by goals for the home team and goals for the away team. We want to visualize the total number of goals scored in each match.


```python
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
df.head()
```

## Bar Chart Visualisation

We are going to create a bar chart visualizing how many goals were scored each year the World Cup was held between 1930-2014.

```python
# Draw barplot
sns.set_style('whitegrid')
sns.set_context('poster', font_scale = 0.5)
f, ax = plt.subplots(figsize = (12, 7))
ax = sns.barplot(data = df, x = df['Year'], y = df['Total Goals'])
ax.set_title('Goals Total between 1930-2014')
plt.show()
```

## Box Plot Visualisation

Now we are going to create a box plot to visualize the distribution of the goals data instead of just the average with a bar chart.

```python
# Load goals.csv into a DataFrame 
df_goals = pd.read_csv('goals.csv')
df_goals.head()
```

```python
# Draw boxplot
sns.set_context('notebook', font_scale = 1.25)
f, ax2 = plt.subplots(figsize = (12, 7))
ax2 = sns.boxplot(data = df_goals, x = df_goals['year'], y = df_goals['goals'], palette = 'Spectral')
ax2.set_title('Distribution of the goals data between 1930-2014')
plt.show()
```
