## Data Investigation

We will start by getting a feel for the `stream` dataset and the `chat` dataset via SQL.
To do display and subsequently visualise the data we'll need convert it to pandas DataFrames. So our steps are:
- import necessary tools;
- create an SQL database;
- upload csv files to variables, creating pandas DataFrames out of them;
- and finally, write records of these DataFrames to new tables in the database. 

```python
# Import tools
import pandas
import sqlite3
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np 

# Create new db
conn = sqlite3.connect('./twitch.db')

# Read csv to DataFrame
chat_df = pandas.read_csv('./chat.csv')
# Create new table for chats in db
chat_df.to_sql('chat', conn, if_exists='replace', index=False)

# Read csv to DataFrame
stream_df = pandas.read_csv('./stream.csv')
# Create new table for streams in db
stream_df.to_sql('stream', conn, if_exists='replace', index=False)
# Create db cursor
curs = conn.cursor()
```

Let's see what we have in those tables. To do that though, we'll convert SQL return results back to `DataFrames`, that will allow us to see results in a table view and make visualizations in the future.

```python
def sql(*args, **kwargs):
    v = curs.execute(*args, **kwargs)
    return pandas.DataFrame(v.fetchall(), columns=[column_name[0] for column_name in v.description])
```

```python
# Display first 5 rows of chat table
sql('''
select * 
from chat
limit 5
''')
```

```python
# Display first 5 rows of stream table
sql('''
select *
from stream
limit 5
''')
```

What are the unique games in the stream table?

```python
sql('''
    select distinct game
    from stream
''')
```

## Aggregate Functions Application

Now we are going to apply different kinds of aggregations and other actions to pull out necessary data. 
First out question is what are the most popular games in the stream table? We are going to find it by counting number of viewers. 

```python
most_popular = sql('''
    select game, count(*)
        as "viewers_count"
    from stream
    group by 1
    order by 2 desc
''')
most_popular
```

Most viewers watch League of Legends streams (also known as LoL). Let's find where are these LoL stream viewers located?

```python
whereabouts = sql('''
    select country, count(*)
        as 'viewers_count'
    from stream
    where game = 'League of Legends'
    group by 1
    order by 2 desc
    limit 20
''')
whereabouts
```

The player column contains the source the user is using to view the stream (site, iphone, android, etc). Let's find out how many users have every streaming source. 

```python
sql('''
    select player, count(*)
        as 'user_count'
    from stream
    group by 1
    order by 2 desc
''')
```

There is an codecademy assignment to add genres of the games with the following logic: 

- If League of Legends → MOBA
- If Dota 2 → MOBA
- If Heroes of the Storm → MOBA
- If Counter-Strike: Global Offensive → FPS
- If DayZ → Survival
- If ARK: Survival Evolved → Survival
- Else → Other


```python
sql('''
    select game,
        case 
            when game = 'League of Legends'
                then 'MOBA'
            when game = 'Dota 2'
                then 'MOBA'
            when game = 'Heroes of the Storm'
                then 'MOBA'
            when game = 'Counter-Strike: Global Offensive'
                then 'FPS'
            when game = 'DayZ'
                then 'Survival'
            when game = 'ARK: Survival Evolved'
                then 'Survival'
        else 'Other'
        end as 'genre',
    count(*)
        as 'viewers_count'
    from stream
    group by 1
    order by 3 desc
''')
```

How does view count change in the course of a day?

```python
viewers_byhours = sql('''
    select strftime('%H', time)
        as 'hour_of_day', 
    count(*)
        as 'views_count'
    from stream
    group by 1
    order by 2 desc
''')
viewers_byhours
```

Result: peak hours is 12PM and least busy is 6 AM.
What about user activity in Sweden, where I live? 

```python
viewers_byhours_SE = sql('''
    select strftime('%H', time)
        as 'hour_of_day', 
    count(*)
        as 'views_count'
    from stream
    where country = 'SE'
    group by 1
    order by 2 desc
''')
viewers_byhours_SE
```

Result: peak hours the same 12PM and least busy is almost the same - 7 AM.

Now let's find out how many people in different countries watch a game I played recently: 'The Witcher 3: Wild Hunt'.

```python
sql('''
    select country, count(*)
    from stream
    where game = 'The Witcher 3: Wild Hunt'
    group by 1
    order by 2 desc
''')
```

## Twitch Data Visualization

Now that we have conducted some analysis with SQL, we will be taking our findings from the SQL queries and visualize them using Matplotlib and Seaborn, in the forms of:

- Bar Graph: Featured Games
- Pie Chart: Stream Viewers’ Locations
- Line Graph: Time Series Analysis


### Bar Graph: Featured Games

We are going to take only those games that have more that 5000 viewers.


```python
most_popular_above_5000 = most_popular[most_popular['viewers_count'] > 5000]
most_popular_above_5000
```

```python
# Draw barplot
plt.figure(figsize=(15, 10))
sns.barplot(
    data = most_popular_above_5000, 
    x='game', 
    y='viewers_count', 
    palette=[
        'Maroon', 
        'Brown', 
        'Sienna', 
        'SaddleBrown', 
        'Olive', 
        'Chocolate', 
        'Peru', 
        'DarkGoldenRod', 
        'GoldenRod', 
        'SandyBrown', 
        'RosyBrown'])

# Add descriptions
plt.title('Featured Games Viewers', fontsize = 'xx-large')

plt.xlabel('Games', fontsize = 'large')
plt.ylabel('Viewers Count', fontsize = 'large')

plt.xticks(
    horizontalalignment='right', 
    fontsize = 'large',
    rotation=45)

plt.show()
```

### Pie Chart: League of Legends Viewers' Whereabouts

There are 200 th. League of Legends viewers from this dataset. Where are they coming from?

```python
# Prepare figure
plt.figure(figsize = (18,13))

# Sum smaller values under status "Other"
other_value = whereabouts.viewers_count[6:].sum()
statuses = whereabouts.viewers_count[0:6]
statuses = np.append(statuses, other_value, axis=None)
statuses

# Draw pie chart
plt.pie(
    statuses, 
    labels=[
    'United States',
    'Canada',
    'Germany',
    'Unknown',
    'Great Britain',
    'Turkey',
    'Other'], 
    autopct='%0.1f%%', 
    textprops={'fontsize': 'large'},
    colors= [
        'Brown', 
        'Sienna', 
        'Olive', 
        'Chocolate', 
        'GoldenRod', 
        'SandyBrown', 
        'RosyBrown'],
    explode= (0.06, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05),
    shadow=True, 
    startangle=260)

plt.axis('equal')

# Add title
plt.title('Viewers Location Distribution', fontsize='x-large')

# Set margins
plt.margins(0.08)

plt.show()
```

Result: The majority is coming from the big and wealthy countries like US and Canada.


### Line Graph: Time Series Analysis

We were able to find the number of Sweden viewers at different hours of the day on January 1st, 2015. Let’s make this into a line graph. There is some uncertainty in these numbers because some people leave their browsers open. Let’s account for a 15% error in the `views_count` data.

```python
# Sort hours in ascending order
viewers_byhours_SE = viewers_byhours_SE.sort_values(by='hour_of_day', ascending=True)
# Change hours format to numeric
viewers_byhours_SE.hour_of_day = pandas.to_numeric(viewers_byhours_SE.hour_of_day)
viewers_byhours_SE.dtypes
```

```python
# Prepare figure
plt.figure(figsize = (15,10))
# Draw lineplot
axes = sns.lineplot(data= viewers_byhours_SE, x = 'hour_of_day', y = 'views_count', color='Olive')
# Add 15% error
y_upper = [i + (i*0.15) for i in viewers_byhours_SE.views_count]
y_lower = [i - (i*0.15) for i in viewers_byhours_SE.views_count]
plt.fill_between(viewers_byhours_SE.hour_of_day, y_lower, y_upper, alpha=0.2, color='Olive') 

# Add descriptions
plt.xticks(fontsize=12)
plt.yticks(fontsize=11)  

plt.xlabel('Hours', fontsize=14)
plt.ylabel('Views Count', fontsize=14)
plt.title('Views count in Sweden', fontsize=15)                 

# Expand chart to full canvas 
axes.set_xlim([min(viewers_byhours_SE['hour_of_day']), max(viewers_byhours_SE['hour_of_day'])])
axes.set_ylim([0, max(viewers_byhours_SE['views_count'])*1.3])
```

Let's compare user activity by hours on Sweden and the rest of the world. Considering that number of viewers is on a vastly different scale one should normalise the viewers count first. 

```python
# Normalise world's data
viewers_byhours_s = viewers_byhours.sort_values(by='hour_of_day', ascending=True).reset_index(drop=True)
viewers_byhours_s['views_normalised'] = viewers_byhours_s.views_count / viewers_byhours_s.views_count.sum()
```

```python
# Normalise Sweden's data
viewers_byhours_SE_s = viewers_byhours_SE.sort_values(by='hour_of_day', ascending=True).reset_index(drop=True)
viewers_byhours_SE_s['views_normalised'] = viewers_byhours_SE_s.views_count / viewers_byhours_SE_s.views_count.sum()
```

```python
# Prepare figure
plt.figure(figsize = (15,10))
# Draw lineplot 1
axes1 = sns.lineplot(data= viewers_byhours_SE_s, x = 'hour_of_day', y = 'views_normalised', color='Olive', label='Sweden')
# Add 15% error
y_upper1 = [i + (i*0.15) for i in viewers_byhours_SE_s.views_normalised]
y_lower1 = [i - (i*0.15) for i in viewers_byhours_SE_s.views_normalised]
plt.fill_between(viewers_byhours_SE_s.hour_of_day, y_lower1, y_upper1, alpha=0.2, color='Olive') 

# Draw lineplot 2
axes2 = sns.lineplot(data= viewers_byhours_s, x = 'hour_of_day', y = 'views_normalised', color='Chocolate', label='World')
# Add 15% error
y_upper2 = [i + (i*0.15) for i in viewers_byhours_s.views_normalised]
y_lower2 = [i - (i*0.15) for i in viewers_byhours_s.views_normalised]
plt.fill_between(viewers_byhours_s.hour_of_day, y_lower2, y_upper2, alpha=0.2, color='Chocolate') 

# Add descriptions
plt.xticks(fontsize=12)
plt.yticks(fontsize=11)     

plt.xlabel('Hours', fontsize=14)
plt.ylabel('Views Count', fontsize=14)
plt.title('Views count in Sweden VS rest of the world', fontsize=15)
plt.legend(fontsize=12)

# Expand charts to full canvas 
axes1.set_xlim([min(viewers_byhours_SE_s['hour_of_day']), max(viewers_byhours_SE_s['hour_of_day'])])
axes1.set_ylim([0, max(viewers_byhours_SE_s['views_normalised'])*2.3])

axes2.set_xlim([min(viewers_byhours_s['hour_of_day']), max(viewers_byhours_s['hour_of_day'])])
axes2.set_ylim([0, max(viewers_byhours_s['views_normalised'])*2.3])
```
